```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
swedish = pd.read_csv("data/swedish_all.csv")
brca = pd.read_csv("data/TCGA_source/tcga_all_brca.csv")
```

```python
swedish.drop("expert_PAM50_subtypes", axis=1, inplace=True)
brca.drop(["sample", "tumor_type", "project", "tissue_site", "participant", "cancer_code", "sample_type"], axis=1, inplace=True)
```

```python
swedish_index = swedish.columns.tolist()
brca_index=brca.columns.tolist()
common_indices = list(set(swedish_index).intersection(brca_index))
```

```python
len(common_indices)
```

```python
swedish_common = swedish[common_indices]
brca_common = brca[common_indices]
```

### Make sure that both data sets are log2(x+1) transformed

```python
swedish_common_transformed = swedish_common.apply(lambda x: 2**x-0.1).apply(lambda x: np.log2(x+1))
```

## Plot the distribution of all genes for both data sets

```python
swedish_common_transformed.mean().hist(bins=100, figsize=(30, 10))
plt.xlim(0, 18)
plt.ylim(0,2700)
```

```python
brca_common.mean().hist(bins=100, figsize=(30, 10))
plt.xlim(0, 18)
plt.ylim(0,2700)
```

# Load and plot Ciriello data set to compare with original tcga - see if the original is upper quartile normalized

```python
brca_ciriello = pd.read_csv("data/ciriello_brca_train.csv")
```

```python
brca_ciriello.mean().hist(bins=100, figsize=(30, 10))
plt.xlim(0, 18)
plt.ylim(0,2700)
```

### From the similarity of the distributions, original tcga are likely to be upper quartile normalized as well

## Now we select randomly 50 genes in order to compare the distribution of its values on both data sets

```python
gene_set = np.random.choice(common_indices, 50, replace=False).tolist()
```

```python
swedish_subset = swedish_common_transformed[gene_set]
brca_subset = brca[gene_set]
```

```python
swedish_subset.boxplot(column=gene_set, figsize=(30, 5))
plt.ylim(-1, 18)
```

```python
brca_subset.boxplot(column=gene_set, figsize=(30, 5))
plt.ylim(-1, 18)
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
swedish_subset.describe()
```

```python
brca_subset.describe()
```

## Try to upper quantile normalize Swedish data in order to compare them in a better way

```python
swedish_upper_quantile = swedish_common_transformed.quantile(0.75)
```

```python
swedish_common_normalized = swedish_common_transformed.div(swedish_upper_quantile, axis=1)
```

# Now we plot again both histograms

```python
swedish_common_normalized
```

```python
swedish_common_normalized.mean().hist(bins=100, figsize=(30, 10))
plt.xlim(0, 18)
plt.ylim(0,2700)
```
