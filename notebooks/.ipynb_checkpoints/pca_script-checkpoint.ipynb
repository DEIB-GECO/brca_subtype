```python
import pandas as pd
```

```python
from sklearn.decomposition import PCA
```

```python
import numpy as np
```

```python
data = pd.read_csv("../data/processed/renamed_TCGA817-FULL_SET.csv")
```

```python
data
```

```python
data = data.drop("TCGA_ID", axis="columns")
```

```python
labels = data["Ciriello_subtype"]
```

```python
labels
```

```python
data = data.drop("Ciriello_subtype", axis="columns")
```

```python
data
```

```python
pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(data)
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, labels], axis="columns")
```

```python
finalDf
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Normal', 'Basal', 'LumA', 'LumB', 'Her2']
colors = ['r', 'orange', 'k', 'g', 'c']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Ciriello_subtype'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
