```python
import pandas as pd
import numpy as np
```

```python
archs4_brca = pd.read_pickle("../data/archs4_brca_19036_row_log_norm.pkl")
```

```python
archs4_mammary = pd.read_csv("../data/archs4_mammary.tsv", sep="\t", index_col=0).T
```

```python
archs4_mammary = archs4_mammary[archs4_brca.columns.values]
```

```python
archs4_mammary_1000000 = archs4_mammary*1000*1000
```

```python
archs4_mammary_norm = archs4_mammary_1000000.div(archs4_mammary.sum(axis=1), axis=0)
```

```python
archs4_mammary_log_norm = archs4_mammary_norm.apply(lambda x: np.log2(x+1))
```

```python
archs4_mammary_log_norm.to_pickle("../data/archs4_mammary_19036_row_log_norm.pkl")
```

## Normalize and merge cell lines data

```python
archs4_cell_MCF7 = pd.read_csv("../data/archs4_cell_line_MCF7.tsv", sep="\t", index_col=0).T
archs4_cell_MCF10 = pd.read_csv("../data/archs4_cell_line_MCF10.tsv", sep="\t", index_col=0).T
archs4_cell_MDAMB231 = pd.read_csv("../data/archs4_cell_line_MDAMB231.tsv", sep="\t", index_col=0).T
archs4_cell_MDAMB468 = pd.read_csv("../data/archs4_cell_line_MDAMB468.tsv", sep="\t", index_col=0).T
archs4_cell_SKBR3 = pd.read_csv("../data/archs4_cell_line_SKBR3.tsv", sep="\t", index_col=0).T
archs4_cell_T47D = pd.read_csv("../data/archs4_cell_line_T47D.tsv", sep="\t", index_col=0).T
```

```python
archs4_cell_all
```

```python
archs4_cell_MCF7 = archs4_cell_MCF7[archs4_brca.columns.values]
archs4_cell_MCF10 = archs4_cell_MCF10[archs4_brca.columns.values]
archs4_cell_MDAMB231 = archs4_cell_MDAMB231[archs4_brca.columns.values]
archs4_cell_MDAMB468 = archs4_cell_MDAMB468[archs4_brca.columns.values]
archs4_cell_SKBR3 = archs4_cell_SKBR3[archs4_brca.columns.values]
archs4_cell_T47D = archs4_cell_T47D[archs4_brca.columns.values]
```

```python
archs4_cell_all = pd.concat([archs4_cell_MCF7, archs4_cell_MCF10, archs4_cell_MDAMB231, archs4_cell_MDAMB468, archs4_cell_SKBR3, archs4_cell_T47D])
```

```python
archs4_cell_all_1000000 = archs4_cell_all*1000*1000
```

```python
archs4_cell_all_norm = archs4_cell_all_1000000.div(archs4_cell_all.sum(axis=1), axis=0)
```

```python
archs4_cell_all_log_norm = archs4_cell_all_norm.apply(lambda x: np.log2(x+1))
```

```python
archs4_cell_all_log_norm.to_pickle("../data/archs4_cell_lines_19036_row_log_norm.pkl")
```

## Now that we normalized all data, let's do the PCA analysis

```python
archs4_brca = pd.read_pickle("../data/archs4_brca_19036_row_log_norm.pkl")
archs4_mammary = pd.read_pickle("../data/archs4_mammary_19036_row_log_norm.pkl")
archs_4_cell_all = pd.read_pickle("../data/archs4_cell_lines_19036_row_log_norm.pkl")
```

```python
archs4_brca_mammary = pd.concat([archs4_brca, archs4_mammary])
```

### We'll add TCGA brca to the analysis as well

```python
tcga_raw_brca_train = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_train.pkl")
tcga_raw_brca_test = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_test.pkl")

tcga_raw_brca_train.drop(["tcga_id", "Ciriello_subtype", "sample_id", "cancer_type"], axis=1, inplace=True)
tcga_raw_brca_test.drop(["tcga_id", "subtype", "sample_id", "cancer_type"], axis=1, inplace=True)

tcga_raw_brca = pd.concat([tcga_raw_brca_train, tcga_raw_brca_test], axis=0)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

archs4_brca_mammary_comp = pca.fit_transform(archs4_brca_mammary)
archs_4_cell_all_comp = pca.fit_transform(archs_4_cell_all)
tcga_brca_comp = pca.fit_transform(tcga_raw_brca)
```

```python
archs4_brca_mammary_comp_df = pd.DataFrame(archs4_brca_mammary_comp, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
archs_4_cell_all_comp_df = pd.DataFrame(archs_4_cell_all_comp, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
tcga_brca_comp_df = pd.DataFrame(tcga_brca_comp, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
archs_4_cell_all.shape
```

## First 2 components PCA

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

brca_mam = ax.scatter(archs4_brca_mammary_comp_df['principal component 1'] , archs4_brca_mammary_comp_df['principal component 2'], c = 'r' , s = 50)
cell_all = ax.scatter(archs_4_cell_all_comp_df['principal component 1'] , archs_4_cell_all_comp_df['principal component 2'], c = 'g' , s = 50)
tcga_brca = ax.scatter(tcga_brca_comp_df['principal component 1'] , tcga_brca_comp_df['principal component 2'], c = 'b' , s = 50)

ax.legend((brca_mam, cell_all, tcga_brca),
           ('ARCHS4 BRCA MAM', 'ARCHS4 Cell line', 'TCGA BRCA'),
           scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=8)
ax.grid()
```

### Now we plot components 1 and 3

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 3', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

brca_mam = ax.scatter(archs4_brca_mammary_comp_df['principal component 1'] , archs4_brca_mammary_comp_df['principal component 3'], c = 'r' , s = 50)
cell_all = ax.scatter(archs_4_cell_all_comp_df['principal component 1'] , archs_4_cell_all_comp_df['principal component 3'], c = 'g' , s = 50)
tcga_brca = ax.scatter(tcga_brca_comp_df['principal component 1'] , tcga_brca_comp_df['principal component 3'], c = 'b' , s = 50)

ax.legend((brca_mam, cell_all, tcga_brca),
           ('ARCHS4 BRCA MAM', 'ARCHS4 Cell line', 'TCGA BRCA'),
           scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=8)
ax.grid()
```

### At last components 2 and 3 

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 2', fontsize = 15)
ax.set_ylabel('Principal Component 3', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

brca_mam = ax.scatter(archs4_brca_mammary_comp_df['principal component 2'] , archs4_brca_mammary_comp_df['principal component 3'], c = 'r' , s = 50)
cell_all = ax.scatter(archs_4_cell_all_comp_df['principal component 2'] , archs_4_cell_all_comp_df['principal component 3'], c = 'g' , s = 50)
tcga_brca = ax.scatter(tcga_brca_comp_df['principal component 2'] , tcga_brca_comp_df['principal component 3'], c = 'b' , s = 50)

ax.legend((brca_mam, cell_all, tcga_brca),
           ('ARCHS4 BRCA MAM', 'ARCHS4 Cell line', 'TCGA BRCA'),
           scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=8)
ax.grid()
```
