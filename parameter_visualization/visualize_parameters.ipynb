# Notebook to plot and visualize the different parameters in order to select the best ones for each model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.legend import Legend
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
```

```python
l_rates = [0.0005, 0.001, 0.01]
b_sizes = [50, 100, 150, 200]
epochs = [25, 50, 75, 100]
```

## Intermediate_dim=100 and latent_dim=100 (in->100->100->100->out)

```python
labels=["learning_rate: 0.0005", "learning_rate: 0.001", "learning_rate: 0.01"]
marker_dict={"0.01":"o", "0.001":"^", "0.0005":"s"}
color_dict = {"50":"firebrick", "100":"seagreen", "150":"sandybrown", "200":"cornflowerblue"}

fig, axes = plt.subplots(figsize=(20, 20))

for l_rate in l_rates:
    for batch in b_sizes:
        for ep in epochs:
            file = pd.read_csv("../parameter_tuning/VAE/two_hidden_100_100/paramsweep_100emb_"+str(l_rate)+"lr_"+str(batch)+"bs_"+str(ep)+"epoch.csv")
            entry = file.tail(1)
            axes.scatter(entry["epochs"].values, 
                         entry["val_loss"].values, 
                         marker=marker_dict[str(l_rate)], 
                         c=color_dict[str(batch)], 
                         s=500, 
                         label="learning_rate: "+str(l_rate)+"\nbatch_size: "+str(batch))
            
figsize=(20,20)
plt.xticks([25, 50, 75, 100])
plt.tick_params(labelsize=30)
plt.grid()

axes.set_ylabel("Final Validation Loss", fontsize=40)
axes.set_xlabel("Epochs", fontsize=40)
axes.set_title("Two Hidden layers: 100->100", fontsize=40)
axes.title.set_position([.5, 1.02])
l_rate_1_patch = mlines.Line2D([0], [0], marker=marker_dict['0.01'], color='w', label='learning_rate: 0.01',
                          markerfacecolor='black', markersize=35)
l_rate_2_patch = mlines.Line2D([0], [0], marker=marker_dict['0.001'], color='w', label='learning_rate: 0.001',
                          markerfacecolor='black', markersize=35)
l_rate_3_patch = mlines.Line2D([0], [0], marker=marker_dict['0.0005'], color='w', label='learning_rate: 0.0005',
                          markerfacecolor='black', markersize=35)
b_size_1_patch = mpatches.Patch(color=color_dict['50'], label='batch_size: 50')
b_size_2_patch = mpatches.Patch(color=color_dict['100'], label='batch_size: 100')
b_size_3_patch = mpatches.Patch(color=color_dict['150'], label='batch_size: 150')
b_size_4_patch = mpatches.Patch(color=color_dict['200'], label='batch_size: 200')
axes.legend(handles=[l_rate_1_patch, 
                     l_rate_2_patch, 
                     l_rate_3_patch, 
                     b_size_1_patch, 
                     b_size_2_patch, 
                     b_size_3_patch, 
                     b_size_4_patch], bbox_to_anchor=(1.6, 0.9), fontsize=40)

fig.savefig('../figures/two_hidden_100_100.png', bbox_inches="tight", pad_inches=1)
```

### Check loss curve for best parameters (b_size=50, l_rate=0.001, epochs=100)

```python
fig, axes = plt.subplots(figsize=(20, 20))

file = pd.read_csv("../parameter_tuning/VAE/two_hidden_100_100/paramsweep_100emb_0.001lr_50bs_100epoch.csv")   
history_df = pd.DataFrame(file[["loss", "val_loss"]])
ax = history_df.plot()
ax.set_xlabel('Epochs')
ax.set_ylabel('VAE Loss')
ax.get_figure()
```
